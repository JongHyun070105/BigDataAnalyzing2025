## Code

```python
# df에 상가상권정보 파일 불러와서 저장하기
df = pd.read_csv('./소상공인시장진흥공단_상가(상권)정보_서울_202412.csv')
df
```

```python
# 상권업종대분류명 컬럼의 유일값 확인하기
# 변수명['컬럼명'].unique()
df['상권업종대분류명'].unique()
```

```python
# 상권업종대분류명이 교육인 데이터만 가지고 와서 복사해서 study에 저장 후 출력
study = df[df['상권업종대분류명'] == '교육']
study
```

```python
# study 데이터의 컬럼별 결측치 값을 n 변수에 저장하고 출력하기
n = study.isna().sum()
n
```

```python
# 결측치가 저장되어 있는 변수 n 값을 내림차순으로 정렬하여
# 상위 10개 컬럼을 not_use 변수에 저장하고 출력하기
not_use = n.sort_values(ascending=False).head(10).index
not_use
```

```python
# 삭제 전 데이터 행, 열 갯수 확인
study.shape
```

```python
# 변수명.drop(삭제할 데이터, 삭제할 데이터 축)
study.drop(not_use, axis = 1, inplace = True)
```

```python
study.shape
```

```python
# study 컬럼명에서 코드 또는 번호 또는 번지가 포함된 컬럼 이름을
# not_use_cols에 저장하고 출력하기
not_use_cols = study.columns[study.columns.str.contains('코드|번지|번호')]
not_use_cols
```

```python
study.drop(not_use_cols, axis = 1, inplace=True)
```

```python
study
```

```python
# study 데이터의 상권업종중분류명 기준으로 유일값 확인하기
study['상권업종중분류명'].unique()
```

```python
# study 데이터의 상권업종소분류명 기준으로 유일값 확인하기
study['상권업종소분류명'].unique()
```

```python
# study 데이터의 상권업종소분류명의 유일값 개수 알아보기
# pandas
study['상권업종소분류명'].nunique()
```

```python
# study 데이터의 상권업종소분류명의 유일값 개수 알아보기
# python
len(study['상권업종소분류명'].unique())
```

```python
# study 데이터의 상권업종소분류명 빈도수 확인하기
study['상권업종소분류명'].value_counts()
```

```python
# study 데이터의 상권업종소분류명 빈도수를 study_count에 저장
study_count = study['상권업종소분류명'].value_counts()
```

```python
# study_count 데이터의 상권업종소분류명 빈도수가 1000개 이상인 데이터만
# study_count_1000에 저장하기

study_count_1000 = study_count[study_count >=1000]
study_count_1000
```

```python
# study 데이터 그룹화하기
# 시군구명과 상권업종소분류명을 기준으로 멀티 인덱스 설정
# 상호명을 기준으로 갯수 세주기
# 결과를 study_group에 저장하고 출력
study_group = study.groupby(['시군구명','상권업종소분류명'])['상호명'].count()
```

```python
# 피봇데이터
# object 형 연산 가능 함수 : count, sum, first, last, min, max
# 시군구명과 상권업종소분류명을 기준으로 멀티 인덱스 설정
# 상호명을 기준으로 갯수 세주기
# 시군구명 - 오름차순, 상호명(갯수) - 내림차순으로 정렬
study.pivot_table(index=['시군구명','상권업종소분류명'], aggfunc='count', values = ['상호명']).sort_values(by=['시군구명','상호명'], ascending=[True, False])
```

```python
# study 데이터의 시군구명 기준으로 빈도수 정렬
study['시군구명'].value_counts()
```

```python
# study 데이터에서 study_count_1000 인 변수를 이용해서
# 상권업종소분류명이 1000개 이상인 데이터만 가져오기
# 데이터프레임.isin(데이터) -> isin 안에 들어간 데이터가 데이터 프레임에 있는지 확인
# 있으면 True, 없으면 False로 출력
study[study['상권업종소분류명'].isin(study_count_1000.index)]
```

```python
# study 데이터에서 study_count_1000 인 변수를 이용해서
# 상권업종소분류명이 1000개 이상인 데이터만 가져오기
# study_selected에 저장 후 출력
study_selected = study[study['상권업종소분류명'].isin(study_count_1000.index)]
study_selected
```

```python
study_selected[study_selected['시군구명'] == '은평구'].head(10)
```

```python
# study_selected 데이터에서 시군구명이 강남구인 법정동명을 확인하기
study_selected[study_selected['시군구명'] == '강남구']['법정동명'].value_counts()
```

```python
# study_selected 를 이용해서 그룹화시키기
# 시군구명, 법정동명, 상권업종소분류명으로 그룹화
# 상호명을 기준으로 갯수세서 study_g에 저장

study_g = study_selected.groupby(by = ['시군구명','법정동명','상권업종소분류명'])['상호명'].count()
study_g
```
